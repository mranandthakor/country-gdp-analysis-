```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#we will use seaborn for stastical visualisation and advance visuallisation like distribition plot...line plot
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.plot(df['IncomeGroup'],c='Green',ls='--',marker='o',ms=4, label='InternetUsers')
```

```python
vis1=sns.distplot(df['InternetUsers'])#plot the graph using one variable called univariate analysis
vis1 #distplot means it show the visualization with distribution lines if we want without distribution line then we will use displot
#plot the graph using two variable called bivariate analysis 
#plot the graph using threee vatiable called multivatiate analysis 
```

```python
sns.displot(df['InternetUsers'])
```

```python
sns.distplot(df['InternetUsers'],bins=20)
```

```python
# if we want to plot two variable we will use box plot
sns.boxplot(data=df,x='IncomeGroup',y='BirthRate')#by default it will print same color boxplot
plt.show()
```

```python
sns.boxplot(data=df,x='IncomeGroup',y='InternetUsers',hue='IncomeGroup')
plt.legend()
plt.show()
```

```python
#let we plot linear line 
sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=True)
```

```python
#if we dont want regression line
sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=False)
```

```python
#for diffrent color we will use hue
sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=True,hue='IncomeGroup')
```
